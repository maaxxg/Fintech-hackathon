# 02 — Churn Label Derivation

Derives binary churn labels for **train + test** clients only (holdout stays label-free).

Three conditions, any one sufficient:
1. **Product closure** — all primary products (excl. DEPOZITI, KANALI) closed in outcome window with no new ones opened.
2. **Transaction cliff** — outcome-window tx count < 15 % of average 4-month tx count in feature window.
3. **Balance collapse** — total balance at T_end < 5 % of average feature-window balance.

Target churn rate: 8–15 %. Thresholds are tuned if outside that range.

**Output:** `data/processed/labels.parquet` — columns: `IDENTIFIKATOR_KLIJENTA`, `churned`

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))
```

```python
import json
from datetime import timezone

import numpy as np
import pandas as pd

import config
```

## 1. Load helpers and raw tables

```python
def parse_datetime(s):
    s = s.astype(str).str.strip().str.replace(r'(\d{4})/', r'\1', regex=True)
    return pd.to_datetime(s, format='%d/%m/%Y %H:%M:%S', errors='coerce', utc=True)


def parse_date_only(s):
    s = s.astype(str).str.strip().str.replace(r'(\d{4})/', r'\1', regex=True)
    return pd.to_datetime(s, format='%d/%m/%Y', errors='coerce', utc=True)


def read_transactions(path):
    cols = [
        'IDENTIFIKATOR_PROIZVODA',
        'DATUM_I_VRIJEME_TRANSAKCIJE',
        'IZNOS_TRANSAKCIJE_U_DOMICILNOJ_VALUTI',
        'VALUTA_TRANSAKCIJE',
        'KANAL',
        'SMJER',
        'DRZAVA_DRUGE_STRANE',
        'DJELATNOST_DRUGE_STRANE',
        'KATEGORIJA_DJELATNOSTI_DRUGE_STRANE',
        'VRSTA_TRANSAKCIJE',
    ]
    df = pd.read_csv(
        path, sep=';', encoding='latin-1',
        names=cols + ['EXTRA'], header=0, low_memory=False,
    )
    mask = df['EXTRA'].notna()
    df.loc[mask, 'KATEGORIJA_DJELATNOSTI_DRUGE_STRANE'] = (
        df.loc[mask, 'KATEGORIJA_DJELATNOSTI_DRUGE_STRANE'].astype(str)
        + ';'
        + df.loc[mask, 'VRSTA_TRANSAKCIJE'].astype(str)
    )
    df.loc[mask, 'VRSTA_TRANSAKCIJE'] = df.loc[mask, 'EXTRA']
    return df.drop(columns=['EXTRA'])
```

```python
RAW = config.RAW_DATA_DIR

# --- Products ---
products = pd.read_csv(RAW / 'HACKATHON ZICER 202604 PROIZVODI.csv', sep=';', encoding='latin-1', low_memory=False)
for col in ['DATUM_OTVARANJA', 'DATUM_ZATVARANJA']:
    products[col] = parse_date_only(products[col])
print(f'Products : {products.shape}')

# --- Transactions (only columns we need) ---
tx = read_transactions(RAW / 'HACKATHON ZICER 202604 TRANSAKCIJE.csv')
tx['DATUM_I_VRIJEME_TRANSAKCIJE'] = parse_datetime(tx['DATUM_I_VRIJEME_TRANSAKCIJE'])
tx['IZNOS_TRANSAKCIJE_U_DOMICILNOJ_VALUTI'] = (
    tx['IZNOS_TRANSAKCIJE_U_DOMICILNOJ_VALUTI']
    .astype(str).str.replace(',', '.', regex=False)
    .pipe(pd.to_numeric, errors='coerce')
)
print(f'Txns     : {tx.shape}')

# --- Balance history ---
balances = pd.read_csv(RAW / 'HACKATHON ZICER 202604 STANJA PROIZVODA.csv', sep=';', encoding='latin-1', low_memory=False)
balances['STANJE_U_DOMICILNOJ_VALUTI'] = (
    balances['STANJE_U_DOMICILNOJ_VALUTI']
    .astype(str).str.replace(',', '.', regex=False)
    .pipe(pd.to_numeric, errors='coerce')
)
balances['VRIJEDI_OD'] = parse_date_only(balances['VRIJEDI_OD'])
balances['VRIJEDI_DO'] = parse_date_only(balances['VRIJEDI_DO'])
print(f'Balances : {balances.shape}')
```

## 2. Load split files and time windows

```python
PROC = config.PROCESSED_DATA_DIR

with open(PROC / 'split_train.json')   as f: train_ids   = set(json.load(f))
with open(PROC / 'split_test.json')    as f: test_ids    = set(json.load(f))
with open(PROC / 'time_split.json')    as f: ts          = json.load(f)

label_ids = train_ids | test_ids  # holdout excluded

T_end               = pd.Timestamp(ts['T_end'])
feature_window_end  = pd.Timestamp(ts['feature_window_end'])
T0                  = pd.Timestamp(ts['T0'])
feature_window_days = (feature_window_end - T0).days

print(f'Labelling clients    : {len(label_ids):,}  (train {len(train_ids):,} + test {len(test_ids):,})')
print(f'Feature window       : {T0.date()} → {feature_window_end.date()}  ({feature_window_days} days)')
print(f'Outcome window       : {feature_window_end.date()} → {T_end.date()}  ({(T_end - feature_window_end).days} days)')
```

## 3. Prepare product-to-client mapping

```python
# Map product → client (for transaction joins)
prod_to_client = (
    products[['IDENTIFIKATOR_PROIZVODA', 'IDENTIFIKATOR_KLIJENTA']]
    .drop_duplicates()
)

# Filter to label clients only
prod_label = products[products['IDENTIFIKATOR_KLIJENTA'].isin(label_ids)].copy()
print(f'Products for label clients: {len(prod_label):,}')
print(f"Domains: {prod_label['NAZIV_DOMENE_PROIZVODA'].value_counts().to_dict()}")
```

## 4. Condition 1 — Product closure

A client is churned if:
- All *primary* products (domains: KARTICE, RAČUNI, KREDITI — excludes DEPOZITI and KANALI) have `DATUM_ZATVARANJA` within the outcome window
- AND no new primary product was opened in the outcome window

```python
PRIMARY_DOMAINS = {'KARTICE', 'RAÈUNI', 'KREDITI'}

# Primary products for label clients
primary = prod_label[prod_label['NAZIV_DOMENE_PROIZVODA'].isin(PRIMARY_DOMAINS)].copy()
print(f'Primary products: {len(primary):,}')

# Per client: how many primary products exist total
total_primary = primary.groupby('IDENTIFIKATOR_KLIJENTA').size().rename('n_primary_total')

# Closed in outcome window: DATUM_ZATVARANJA is not null and falls in [feature_window_end, T_end]
closed_in_outcome = primary[
    primary['DATUM_ZATVARANJA'].notna()
    & (primary['DATUM_ZATVARANJA'] >= feature_window_end)
    & (primary['DATUM_ZATVARANJA'] <= T_end)
].groupby('IDENTIFIKATOR_KLIJENTA').size().rename('n_primary_closed_outcome')

# Newly opened in outcome window (any domain — if they opened a new one they're not churning)
opened_in_outcome = prod_label[
    prod_label['DATUM_OTVARANJA'].notna()
    & (prod_label['DATUM_OTVARANJA'] >= feature_window_end)
    & (prod_label['DATUM_OTVARANJA'] <= T_end)
].groupby('IDENTIFIKATOR_KLIJENTA').size().rename('n_opened_outcome')

# Assemble per-client product closure DataFrame
closure_df = pd.DataFrame({'client': list(label_ids)}).set_index('client')
closure_df = closure_df.join(total_primary).join(closed_in_outcome).join(opened_in_outcome)
closure_df['n_primary_total'].fillna(0, inplace=True)
closure_df['n_primary_closed_outcome'].fillna(0, inplace=True)
closure_df['n_opened_outcome'].fillna(0, inplace=True)

# Condition: all primary products closed AND no new ones opened
cond1 = (
    (closure_df['n_primary_total'] > 0)
    & (closure_df['n_primary_closed_outcome'] >= closure_df['n_primary_total'])
    & (closure_df['n_opened_outcome'] == 0)
)

print(f'Condition 1 (product closure) triggers: {cond1.sum():,} / {len(cond1):,}')
```

## 5. Condition 2 — Transaction cliff

Outcome-window tx count < 15 % of average 4-month tx count in feature window.

```python
# Add client ID to transactions
tx_c = tx.merge(prod_to_client, on='IDENTIFIKATOR_PROIZVODA', how='left')
tx_c = tx_c[tx_c['IDENTIFIKATOR_KLIJENTA'].isin(label_ids)]

# Feature-window transactions
tx_feat = tx_c[
    tx_c['DATUM_I_VRIJEME_TRANSAKCIJE'].notna()
    & (tx_c['DATUM_I_VRIJEME_TRANSAKCIJE'] >= T0)
    & (tx_c['DATUM_I_VRIJEME_TRANSAKCIJE'] < feature_window_end)
]
tx_feat_count = tx_feat.groupby('IDENTIFIKATOR_KLIJENTA').size().rename('tx_feat_count')

# Outcome-window transactions
tx_out = tx_c[
    tx_c['DATUM_I_VRIJEME_TRANSAKCIJE'].notna()
    & (tx_c['DATUM_I_VRIJEME_TRANSAKCIJE'] >= feature_window_end)
    & (tx_c['DATUM_I_VRIJEME_TRANSAKCIJE'] <= T_end)
]
tx_out_count = tx_out.groupby('IDENTIFIKATOR_KLIJENTA').size().rename('tx_out_count')

tx_df = pd.DataFrame({'client': list(label_ids)}).set_index('client')
tx_df = tx_df.join(tx_feat_count).join(tx_out_count)
tx_df['tx_feat_count'].fillna(0, inplace=True)
tx_df['tx_out_count'].fillna(0, inplace=True)

# Average 4-month tx count in feature window
n_4month_periods = feature_window_days / 120.0
tx_df['avg_4month_tx_feat'] = tx_df['tx_feat_count'] / n_4month_periods

# Cliff condition: outcome tx < 15% of avg 4-month feature-window tx
cond2 = (
    (tx_df['avg_4month_tx_feat'] > 0)
    & (tx_df['tx_out_count'] < config.CHURN_TX_CLIFF * tx_df['avg_4month_tx_feat'])
)

print(f'Feature window: {feature_window_days} days ({n_4month_periods:.2f} × 4-month periods)')
print(f'Condition 2 (tx cliff) triggers     : {cond2.sum():,} / {len(cond2):,}')
```

## 6. Condition 3 — Balance collapse

Total balance at T_end < 5 % of average feature-window balance.

```python
# Map balance records to clients
bal_c = balances.merge(
    prod_to_client[prod_to_client['IDENTIFIKATOR_KLIJENTA'].isin(label_ids)],
    on='IDENTIFIKATOR_PROIZVODA', how='inner'
)

# Feature-window balances: records valid during feature window
# A record is "in" the feature window if VRIJEDI_OD < feature_window_end
# (we use the start-of-validity date as the snapshot timestamp)
bal_feat = bal_c[
    bal_c['VRIJEDI_OD'].notna()
    & (bal_c['VRIJEDI_OD'] >= T0)
    & (bal_c['VRIJEDI_OD'] < feature_window_end)
]
avg_bal_feat = (
    bal_feat.groupby('IDENTIFIKATOR_KLIJENTA')['STANJE_U_DOMICILNOJ_VALUTI']
    .mean()
    .rename('avg_bal_feat')
)

# Balance at T_end: most recent balance record at or before T_end
bal_at_tend = (
    bal_c[bal_c['VRIJEDI_OD'].notna() & (bal_c['VRIJEDI_OD'] <= T_end)]
    .sort_values('VRIJEDI_OD')
    .groupby(['IDENTIFIKATOR_KLIJENTA', 'IDENTIFIKATOR_PROIZVODA'])['STANJE_U_DOMICILNOJ_VALUTI']
    .last()  # latest per product
    .reset_index()
    .groupby('IDENTIFIKATOR_KLIJENTA')['STANJE_U_DOMICILNOJ_VALUTI']
    .sum()
    .rename('bal_at_tend')
)

bal_df = pd.DataFrame({'client': list(label_ids)}).set_index('client')
bal_df = bal_df.join(avg_bal_feat).join(bal_at_tend)

# Only trigger when client actually has balance data
cond3 = (
    bal_df['avg_bal_feat'].notna()
    & (bal_df['avg_bal_feat'] > 0)
    & (
        bal_df['bal_at_tend'].fillna(0)
        < config.CHURN_BALANCE_COLLAPSE * bal_df['avg_bal_feat']
    )
)

print(f'Condition 3 (balance collapse) triggers: {cond3.sum():,} / {len(cond3):,}')
```

## 7. Combine conditions and check churn rate

```python
# Align all three Series on the same index
all_clients = pd.Index(list(label_ids))

c1 = cond1.reindex(all_clients, fill_value=False)
c2 = cond2.reindex(all_clients, fill_value=False)
c3 = cond3.reindex(all_clients, fill_value=False)

churned = (c1 | c2 | c3).astype(int)

churn_rate = churned.mean()

print(f'Churned via cond1 only  : {(c1 & ~c2 & ~c3).sum():,}')
print(f'Churned via cond2 only  : {(~c1 & c2 & ~c3).sum():,}')
print(f'Churned via cond3 only  : {(~c1 & ~c2 & c3).sum():,}')
print(f'Churned via 2+ conditions: {(c1.astype(int) + c2.astype(int) + c3.astype(int) >= 2).sum():,}')
print(f'Total churned           : {churned.sum():,} / {len(churned):,}')
print(f'Churn rate              : {churn_rate:.1%}')

if not (0.08 <= churn_rate <= 0.15):
    print(f'\nWARNING: churn rate {churn_rate:.1%} is outside target range 8–15%')
else:
    print('\nChurn rate is within target range (8–15%). OK.')
```

## 8. Save labels.parquet

```python
labels_df = pd.DataFrame({
    'IDENTIFIKATOR_KLIJENTA': churned.index,
    'churned': churned.values,
    'cond1_product_closure':  c1.values,
    'cond2_tx_cliff':         c2.values,
    'cond3_balance_collapse': c3.values,
})

out_path = config.PROCESSED_DATA_DIR / 'labels.parquet'
labels_df.to_parquet(out_path, index=False)
print(f'Saved: {out_path}')
print(f'Shape: {labels_df.shape}')
print(labels_df.head())
```

## 9. Train / test churn rate split check

```python
train_labels = labels_df[labels_df['IDENTIFIKATOR_KLIJENTA'].isin(train_ids)]
test_labels  = labels_df[labels_df['IDENTIFIKATOR_KLIJENTA'].isin(test_ids)]

print(f'Train  — n={len(train_labels):,}, churn rate={train_labels["churned"].mean():.1%}')
print(f'Test   — n={len(test_labels):,}, churn rate={test_labels["churned"].mean():.1%}')
```

## 10. Summary

```python
print('=' * 52)
print('CHURN LABEL SUMMARY')
print('=' * 52)
print(f'Labelled clients total  : {len(labels_df):,}')
print(f'  Churned               : {labels_df["churned"].sum():,}  ({labels_df["churned"].mean():.1%})')
print(f'  Not churned           : {(labels_df["churned"]==0).sum():,}  ({(labels_df["churned"]==0).mean():.1%})')
print()
print(f'Condition triggers:')
print(f'  cond1 (product closure) : {labels_df["cond1_product_closure"].sum():,}')
print(f'  cond2 (tx cliff)        : {labels_df["cond2_tx_cliff"].sum():,}')
print(f'  cond3 (balance collapse): {labels_df["cond3_balance_collapse"].sum():,}')
print('=' * 52)
```
